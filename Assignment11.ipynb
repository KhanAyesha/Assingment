9. What are the differences between the Step Over, Step In, and Step Out buttons in the debugger?

The Step In button will move the debugger into a function call. The Over button will quickly execute the function call without stepping into it. The Out button will quickly execute the rest of the code until it steps out of the function it currently is in.
